**Background:**

The thyroid gland, located in the anterior region of the neck, is responsible for producing and regulating key hormones involved in metabolism and energy balance. These include thyroxine (T4) and triiodothyronine (T3), which are secreted in response to stimulation by thyroid-stimulating hormone (TSH), a hormone produced by the pituitary gland. The balance of these hormones is critical for numerous physiological functions, including:

1.   Regulation of basal metabolic rate
2.   Cardiovascular and gastrointestinal activity
3.   Muscular development and bone turnover
4.   Brain development and neurocognitive functioning

719086-571.jpg

TSH is the most commonly used biomarker to assess thyroid health due to its sensitivity to even slight changes in circulating T3 and T4 levels. Free T4 (FT4) represents the unbound, biologically active form of thyroxine, while total T3 provides an estimate of overall triiodothyronine activity. Clinically, these markers are used to diagnose and monitor thyroid disorders such as:

1. Hypothyroidism, associated with low FT4 and elevated TSH levels
2. Hyperthyroidism, associated with high FT4/T3 and suppressed TSH levels

Even in the absence of overt disease, subclinical abnormalities in thyroid hormone levels can contribute to metabolic disturbances, fatigue, and increased cardiovascular risk. Therefore, understanding patterns in hormone distribution, particularly within specific populations, is important for early identification of thyroid dysfunction and for refining public health interventions.

Prior studies have shown that thyroid hormone concentrations can differ by demographic factors such as age, sex, and race/ethnicity. For example, women are known to have a higher prevalence of autoimmune thyroid disease, while men may present differently and be diagnosed less frequently (Hollowell et al., 2002). Hispanic adults are an understudied group in thyroid epidemiology, and there is limited data describing sex-specific distributions of TSH, Free T4, and Total T3 within this population.

Furthermore, much of the existing evidence is based on older surveys or studies conducted outside the United States. Since the early 2000s, shifts in iodine intake, obesity prevalence, and smoking behaviors may have altered thyroid hormone distributions in the US Hispanic population. NHANES data from 2007–2012 uniquely include standardized laboratory measures of all three classical thyroid markers in a nationally representative sample.

```python
import pandas as pd
import numpy as np

# --------------------------------
# Dataset Configuration
# --------------------------------
years = ["2007", "2009", "2011"]
cycles = ["E", "F", "G"]

modules = {
    "demographics": {
        "prefix": "DEMO",
        "vars": ["SEQN", "RIAGENDR", "RIDAGEYR", "RIDRETH1"]
    },
    "bmi": {
        "prefix": "BMX",
        "vars": ["SEQN", "BMXBMI"]
    },
    "smoking": {
        "prefix": "SMQ",
        "vars": ["SEQN", "SMQ040"]
    },
    "alcohol": {
        "prefix": "ALQ",
        "vars": ["SEQN", "ALQ101"]
    },
    "iodine": {
        "prefix": "UIO",
        "vars": ["SEQN", "URXUIO"]
    },
    "thyroid": {
        "prefix": "THYROD",
        "vars": ["SEQN", "LBXT4F", "LBXTT3", "LBXTSH1"]
    },
    "sleep": {
        "prefix": "SLQ",
        "vars": ["SEQN", "SLD010H"]
    }
}

# --------------------------------
# Load and Merge NHANES Data
# --------------------------------
def load_nhanes(years, cycles):
    dfs = {}
    for label, config in modules.items():
        prefix = config["prefix"]
        vars_needed = config["vars"]
        df_list = []
        for y, c in zip(years, cycles):
            url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/{prefix}_{c}.XPT"
            temp = pd.read_sas(url)[vars_needed]
            df_list.append(temp)
        df = pd.concat(df_list, ignore_index=True)
        dfs[label] = df

    df = dfs["demographics"]
    for label, sub_df in dfs.items():
        if label != "demographics":
            df = df.merge(sub_df, on="SEQN", how="inner")
    return df

# --------------------------------
# Clean and Filter for Hispanic Adults
# --------------------------------
def prepare_dataset(df):
    df = df.copy()

    # Recode Gender and Race
    df["Gender"] = df["RIAGENDR"].map({1: "Male", 2: "Female"})
    df["Race"] = df["RIDRETH1"].map({
        1: "Mexican American",
        2: "Other Hispanic",
        3: "Non-Hispanic White",
        4: "Non-Hispanic Black",
        5: "Other/Multi-Racial"
    })

    # Replace known missing codes with NaN
    df.replace({7777: np.nan, 9999: np.nan, 8888: np.nan, 99999: np.nan}, inplace=True)
    df["SLD010H"] = df["SLD010H"].replace({77: np.nan, 99: np.nan})
    df["SMQ040"] = df["SMQ040"].replace({7: np.nan, 9: np.nan})
    df["ALQ101"] = df["ALQ101"].replace({7: np.nan, 9: np.nan})

    # Drop rows missing hormone or demographic values
    df = df.dropna(subset=["LBXT4F", "LBXTT3", "LBXTSH1", "RIDAGEYR"])

    # Filter: Hispanic adults (≥18 years old)
    df = df[(df["RIDAGEYR"] >= 18) & (df["RIDRETH1"].isin([1, 2]))]

    # Select final columns
    return df[[
        "SEQN", "RIDAGEYR", "Gender", "Race", "BMXBMI", "SMQ040", "ALQ101",
        "URXUIO", "SLD010H", "LBXT4F", "LBXTT3", "LBXTSH1"
    ]]

# --------------------------------
# Run Pipeline
# --------------------------------
df_raw = load_nhanes(years, cycles)
df_clean = prepare_dataset(df_raw)

# df_clean is now ready for analysis
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis, shapiro, f_oneway
import numpy as np  # Needed for handling np.nan

# --------------------------------
# Run Descriptive Stats and Normality Tests
# --------------------------------
def run_stats(df, hormone, group_col):
    print(f"\n📊 {hormone} Analysis by {group_col}:")

    # Frequency table
    freq_table = df[group_col].value_counts(dropna=False)
    print("\nFrequency Table:")
    print(freq_table)

    # Grouped data
    grouped = df.groupby(group_col)[hormone]

    # Descriptive Stats with Standard Deviation
    print("\nDescriptive Statistics (Mean, Median, Std, Q1, Q3, IQR):")
    stats_table = grouped.agg([
        'count', 'mean', 'median', 'std',
        lambda x: x.quantile(0.25),
        lambda x: x.quantile(0.75)
    ])
    stats_table.columns = ["N", "Mean", "Median", "Std", "Q1", "Q3"]
    stats_table["IQR"] = stats_table["Q3"] - stats_table["Q1"]
    print(stats_table)

    # Shapiro-Wilk, Skewness, Kurtosis
    print("\nNormality Tests and Distribution Metrics:")
    for group, data in grouped:
        values = data.dropna()
        if len(values) < 3:
            print(f"{group}: Not enough data (N={len(values)})")
            continue
        s, p = shapiro(values) if len(values) <= 5000 else (np.nan, np.nan)
        print(f"\nGroup: {group}")
        print(f"  N = {len(values)}")
        print(f"  Skewness: {skew(values):.4f}")
        print(f"  Kurtosis: {kurtosis(values):.4f}")
        if not np.isnan(s):
            print(f"  Shapiro-Wilk: W={s:.4f}, p = {p:.4f} →", "Not normal" if p < 0.05 else "Normal")
        else:
            print("  Shapiro-Wilk skipped (N > 5000)")

    # One-way ANOVA
    anova_groups = [g.dropna() for _, g in grouped if len(g.dropna()) >= 3]
    if len(anova_groups) >= 2:
        f, pval = f_oneway(*anova_groups)
        print(f"\nOne-Way ANOVA: F = {f:.4f}, p = {pval:.4f}")
        print("→ Significant difference" if pval < 0.05 else "→ No significant difference")
    else:
        print("\n→ Not enough valid groups for ANOVA")

# --------------------------------
# Plot Histograms and Run Stats for Alcohol
# --------------------------------
def analyze_alcohol_vs_thyroid(df):
    df = df.copy()
    df["Alcohol_Consumption"] = df["ALQ101"].map({1: "Yes", 2: "No"})
    df = df.dropna(subset=["Alcohol_Consumption"])

    hormones = {
        "LBXT4F": "Free T4 (ng/dL)",
        "LBXTT3": "Total T3 (ng/dL)",
        "LBXTSH1": "TSH (uIU/mL)"
    }

    for hvar, hname in hormones.items():
        plt.figure(figsize=(10, 6))
        sns.histplot(data=df, x=hvar, hue="Alcohol_Consumption", kde=True, bins=30, multiple="stack", palette="Set2")
        plt.title(f"{hname} by Alcohol Consumption")
        plt.xlabel(hname)
        plt.ylabel("Frequency")
        plt.tight_layout()
        plt.show()

        run_stats(df, hvar, "Alcohol_Consumption")

# --------------------------------
# Run Analysis
# --------------------------------
analyze_alcohol_vs_thyroid(df_clean)
```

# Annual Trend


---

This section presents annual trends in median concentrations of TSH, Free T4, and Total T3 across the NHANES 2007–2012 cycles. By visualizing median values and the percentage of participants in clinically abnormal ranges, we aim to identify potential shifts in thyroid hormone distributions over time. These trends can provide insight into temporal patterns in thyroid function and inform public health monitoring efforts.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# --- Simulated NHANES-like data (replace with your own data as needed) ---
np.random.seed(0)
n = 300
df_raw = pd.DataFrame({
    'SDDSRVYR': np.random.choice([5,6,7], n),
    'LBXTSH1': np.random.normal(loc=2.5, scale=1.0, size=n),
    'LBXT4F': np.random.normal(loc=1.2, scale=0.3, size=n),
    'LBXTT3': np.random.normal(loc=110, scale=20, size=n)
})

# --- Data cleaning ---
df_clean = df_raw[
    (df_raw['LBXTSH1'] > 0) &
    (df_raw['LBXT4F'] > 0) &
    (df_raw['LBXTT3'] > 0)
].copy()

# Map cycles to readable labels
cycle_map = {5: "2007-2008", 6: "2009-2010", 7: "2011-2012"}
df_clean['Cycle'] = df_clean['SDDSRVYR'].map(cycle_map)

# Create indicator columns BEFORE grouping
df_clean['High_TSH'] = df_clean['LBXTSH1'] > 4.5
df_clean['Low_TSH'] = df_clean['LBXTSH1'] < 0.4

# Compute median hormone levels by cycle
medians = df_clean.groupby('Cycle')[['LBXTSH1', 'LBXT4F', 'LBXTT3']].median().reset_index()

# --- Plot median TSH ---
plt.figure(figsize=(6,4))
plt.plot(medians['Cycle'], medians['LBXTSH1'], marker='o', color='blue')
plt.title("Median TSH by Cycle")
plt.xlabel("NHANES Cycle")
plt.ylabel("Median TSH (uIU/mL)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()

# --- Plot median Free T4 ---
plt.figure(figsize=(6,4))
plt.plot(medians['Cycle'], medians['LBXT4F'], marker='o', color='green')
plt.title("Median Free T4 by Cycle")
plt.xlabel("NHANES Cycle")
plt.ylabel("Median Free T4 (ng/dL)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()

# --- Plot median Total T3 ---
plt.figure(figsize=(6,4))
plt.plot(medians['Cycle'], medians['LBXTT3'], marker='o', color='purple')
plt.title("Median Total T3 by Cycle")
plt.xlabel("NHANES Cycle")
plt.ylabel("Median Total T3 (ng/dL)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()

# --- Calculate % Elevated TSH ---
elevated_counts = df_clean.groupby('Cycle')['High_TSH'].sum()
total_counts = df_clean.groupby('Cycle')['High_TSH'].count()
elevated_pct = (elevated_counts / total_counts * 100).reset_index()
elevated_pct.columns = ['Cycle', 'Elevated_TSH_Pct']

# --- Bar plot of Elevated TSH ---
plt.figure(figsize=(6,5))
plt.bar(elevated_pct['Cycle'], elevated_pct['Elevated_TSH_Pct'], color='salmon')
plt.title("Percentage of Participants with TSH > 4.5 uIU/mL")
plt.xlabel("NHANES Cycle")
plt.ylabel("% Elevated TSH")
plt.ylim(0, max(elevated_pct['Elevated_TSH_Pct'].max() + 2, 5))
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()

# --- Calculate % Low TSH ---
low_counts = df_clean.groupby('Cycle')['Low_TSH'].sum()
low_pct = (low_counts / total_counts * 100).reset_index()
low_pct.columns = ['Cycle', 'Low_TSH_Pct']  # <-- FIXED NAME

# --- Bar plot of Low TSH ---
plt.figure(figsize=(6,5))
plt.bar(low_pct['Cycle'], low_pct['Low_TSH_Pct'], color='skyblue')  # <-- FIXED NAME
plt.title("Percentage of Participants with TSH < 0.4 uIU/mL")
plt.xlabel("NHANES Cycle")
plt.ylabel("% Low TSH")
plt.ylim(0, max(low_pct['Low_TSH_Pct'].max() + 2, 5))
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

**Thyroid Medication**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

years = ["2007", "2009", "2011"]
cycles = ["E", "F", "G"]

modules = {
    "demographics": {
        "prefix": "DEMO",
        "vars": ["SEQN", "RIDAGEYR", "RIDRETH1"]
    },
    "medications": {
        "prefix": "RXQ_RX",
        "vars": ["SEQN", "RXDDRUG"]
    }
}

thyroid_keywords = ["thyroid", "levothyroxine", "liothyronine", "armour"]

def load_cycle_data(year, cycle):
    # Load demographics data
    demo_url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{year}/DataFiles/DEMO_{cycle}.XPT"
    demo = pd.read_sas(demo_url)[modules["demographics"]["vars"]]
    # Filter for Hispanic adults (18+)
    demo = demo[(demo["RIDAGEYR"] >= 18) & (demo["RIDRETH1"].isin([1, 2]))]

    # Load medications data
    meds_url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{year}/DataFiles/RXQ_RX_{cycle}.XPT"
    meds = pd.read_sas(meds_url)[modules["medications"]["vars"]]

    # Decode RXDDRUG if bytes
    if meds["RXDDRUG"].dtype == "O" and isinstance(meds["RXDDRUG"].iloc[0], bytes):
        meds["RXDDRUG"] = meds["RXDDRUG"].apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)

    # Filter thyroid medications by keywords (case-insensitive)
    meds_thyroid = meds[meds["RXDDRUG"].str.lower().str.contains('|'.join(thyroid_keywords), na=False)]

    # Get SEQN for participants on thyroid meds
    thyroid_seqns = set(meds_thyroid["SEQN"])

    # Add flag for thyroid medication use
    demo["Thyroid_Med_Use"] = demo["SEQN"].apply(lambda x: 1 if x in thyroid_seqns else 0)

    return demo

# Collect prevalence per cycle
results = []
for y, c in zip(years, cycles):
    df_cycle = load_cycle_data(y, c)
    prevalence = df_cycle["Thyroid_Med_Use"].mean() * 100  # percent
    results.append({"Cycle": f"{y}-{int(y)+1}", "Thyroid_Med_Use_%": prevalence})

df_results = pd.DataFrame(results)

# Plot thyroid medication use prevalence
plt.figure(figsize=(8,5))
plt.bar(df_results["Cycle"], df_results["Thyroid_Med_Use_%"], color='teal')
plt.ylim(0, df_results["Thyroid_Med_Use_%"].max() + 1)
plt.ylabel("Thyroid Medication Use (%)")
plt.title("Thyroid Medication Use Prevalence in Hispanic Adults (NHANES 2007-2012)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

There is a slight decreasing trend in thyroid medication use among Hispanic adults over the 6-year period.
From 2007–2008 to 2011–2012, the prevalence appears to drop by approximately 0.4 percentage points.

**Interpretation**

- This decline, though modest, may reflect:
- Changes in diagnosis rates
- Access to care
- Population health trends
- Differences in survey sampling or medication adherence
